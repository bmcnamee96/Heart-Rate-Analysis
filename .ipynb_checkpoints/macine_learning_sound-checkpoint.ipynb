```python
# import dependencies
# Python SQL toolkit and Object Relational Mapper
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# import the dataset
file_path = '../Resources/work_sound.csv'
work_df = pd.read_csv(file_path)
work_df.head()
```

```python
ml_df = work_df[['values', 'day_sections', 'day_of_week']]
```

```python
target_df = ml_df[['values']]
feature_df = ml_df[['day_sections', 'day_of_week']]
```

```python
X = feature_df
y = target_df

# create a linear model
model = LinearRegression()
```

```python
# split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# fit our model to the training data
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
X_test['predicted_values'] = y_pred
```

```python
comparison_pd = X_test
```

```python
comparison_pd = comparison_pd.join(y_test)
```

```python
comparison_pd[['predicted_values', 'values']]
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot('predicted_values', 'values', 'o', data=comparison_pd)
plt.show()
```
