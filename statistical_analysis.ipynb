```python
# import dependencies
import pandas as pd

from bioinfokit.analys import stat
```

```python
# load the csv
file_path = '../Resources/work_hr.csv'

# create a df
df = pd.read_csv(file_path)

#show the df
df.head()
```

```python
df['day_sections'].value_counts()
```

```python
res = stat()
res.tukey_hsd(df=df, res_var='values', xfac_var='day_sections', anova_model='values ~ C(day_sections)')
res.tukey_summary
```

```python
res = stat()
res.tukey_hsd(df=df, res_var='values', xfac_var='day_of_week', anova_model='values ~ C(day_of_week)')
res.tukey_summary
```
